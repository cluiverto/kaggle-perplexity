```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import numpy as np
plt.style.use('fivethirtyeight')

import sklearn


train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

test.head()
```

```python
## Missing values

x = train.isna().sum()
items = train.columns
z = round((train[items].isna().sum() / len(train[items])) * 100, 2)
pd.concat([x, z], axis=1, keys=['missing', 'missing_percent'])
```

```python
## Drop missing

train.dropna(inplace=True)
train.isna().sum()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

le = preprocessing.LabelEncoder()

testo = test.copy()
train[['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Transported']] = train[['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Transported']].apply(le.fit_transform)
test[['HomePlanet', 'CryoSleep', 'Destination', 'VIP']] = test[['HomePlanet', 'CryoSleep', 'Destination', 'VIP']].apply(le.fit_transform)
test.drop(columns=['PassengerId', 'Cabin', 'Name'], inplace=True, axis=1)

train
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

# Only floats
float_columns_train = train.select_dtypes(include=['float64']).columns
float_columns_test = train.select_dtypes(include=['float64']).columns


# Scale them
train[float_columns_train] = scaler.fit_transform(train[float_columns_train])
test[float_columns_test] = scaler.fit_transform(test[float_columns_test])


train
```

```python
train.drop(columns=['PassengerId', 'Cabin', 'Name'], inplace=True, axis=1)
```

```python
from sklearn.model_selection import train_test_split

y = train.iloc[:, -1]
X = train.iloc[:, :-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier, GradientBoostingClassifier




dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
y_preds = dt_classifier.predict(X_test)
print(f"Decision Tree Classifier\nTest Accuracy: {accuracy_score(y_test, y_preds)*100}%")
```

```python

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score



# Stworzenie klasyfikatora MLP
mlp_classifier = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000)

# Trenowanie modelu
mlp_classifier.fit(X_train, y_train)

# Predykcja na zbiorze testowym
y_pred = mlp_classifier.predict(X_test)

# Ocena modelu
print("Raport klasyfikacji:")
print(classification_report(y_test, y_pred))
print("Dokładność:", accuracy_score(y_test, y_pred))
```
